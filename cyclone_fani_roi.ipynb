# OSM vector data for ML training 
## Cyclone Fani
So I was curious if i could download OSM buildings and other entities and use it as training data for a machine learning model. In particular, I am really interested in maling a tight, clean pipeline for disaster relief applications. 

An unfortunate, timely cyclone, Fani, has just hit near Puri, India. Digital Globe open data as som post even imagery available. Let's grab OSM vectors for that region. I decided to use the <a href='https://automating-gis-processes.github.io/2017/lessons/L7/retrieve-osm-data.html'>OSMNX</a> python library for interfacing with OSM (which can be a bit daunting otherwise). 



```python
%matplotlib inline
import osmnx as ox
import matplotlib.pyplot as plt

place_name = "Puri,India"

graph = ox.graph_from_place(place_name, which_result=2)
```

```python
type(graph)
fig, ax = ox.plot_graph(graph)
```

```python
area = ox.gdf_from_place(place_name)
buildings = ox.buildings_from_place(place_name)
type(area)
type(buildings)
```
